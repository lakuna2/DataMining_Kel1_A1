####Cross Validation
Validasi silang dilakukan untuk mengevaluasi kinerja model secara objektif dan mencegah overfitting atau underfitting. Pada kasus ini, teknik validasi silang digunakan untuk menentukan nilai k yang optimal dalam membangun model yang lebih akurat.

Output di atas menunjukkan bahwa nilai k yang optimal dalam kasus ini adalah yang bernilai lima belas dengan skor validasi 0.53.

```python
# ambil nilai skor validasi dari grid search
mean_scores = grid_search.cv_results_['mean_test_score']
std_scores = grid_search.cv_results_['std_test_score']
params = grid_search.cv_results_['params']

# plot hasil dari grid search
plt.errorbar(
    [params[i]['n_neighbors'] for i in range(len(params))],
    mean_scores,
    # yerr=std_scores,
    fmt='o-'
)
plt.xlabel('Nilai K')
plt.ylabel('Skor Uji Rata-Rata')
plt.title('Hasil Validasi Silang')
plt.grid(True)
plt.show()
```

#### Naive Bayes
Algoritma ini merupakan metode klasifikasi yang menggunakan probabilitas dan asumsi bahwa setiap fitur input adalah independen satu sama lain, sehingga disebut "naive". Dengan memanfaatkan probabilitas, algoritma ini dapat menghasilkan prediksi rating berdasarkan faktor-faktor yang mempengaruhi secara efisien.

#####Menentukan prior probability:

Setelah menentukan target, kita perlu menentukan prior probability atau probabilitas awal. Hal ini dilakukan dengan menghitung frekuensi kemunculan setiap nilai target pada data training.

```python
class GaussianNB:

  # Menghitung prior probability dari target rating
  prior = dict(y_train.value_counts(normalize=True))

  # Menghitung likelihood probability dari setiap fitur terhadap setiap nilai target rating
  likelihood = {}

  for col in x_train.columns:
      for val in x_train[col].unique():
          for rating in y_train.unique():
              freq = len(x_train[(x_train[col] == val) & (y_train == rating)])
              likelihood[(col, val, rating)] = freq / y_train.value_counts()[rating]

  # Membuat fungsi untuk menghitung probabilitas posterior dari setiap nilai target untuk setiap instance data pada data testing
  def predict(x_test, prior=prior, likelihood=likelihood):
      results = []

      for i in range(len(x_test)):
          p_recommend = {}

          for recommend in y_train.unique():
              p = prior[recommend]

              for col in x_train.columns:
                  val = x_test.iloc[i][col]
                  p *= likelihood.get((col, val, recommend), 1e-6)

              p_recommend[recommend] = p

          results.append(max(p_recommend, key=p_recommend.get))

      return results

  # Memprediksi nilai target pada data testing dengan menggunakan model Naive Bayes
  y_pred = predict(x_test, prior, likelihood)
```

```python
# Menguji model:
# Menghitung akurasi dari model Naive Bayes
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
# Membuat confusion matrix dari model Naive Bayes
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', cm)
```

## Evaluation

Evaluasi pada analisis ini akan dilakukan menggunakan confusion matrix. Confusion matrix merupakan sebuah matriks yang digunakan untuk mengevaluasi kinerja model klasifikasi dengan membandingkan nilai aktual dan prediksi dari suatu model. Terdapat empat nilai dalam confusion matrix, yaitu True Positive (TP), True Negative (TN), False Positive (FP), dan False Negative (FN). Dari nilai-nilai ini, kita dapat menghitung akurasi, presisi, dan recall. Akurasi merupakan rasio prediksi yang benar dari semua prediksi. Presisi merupakan rasio prediksi yang benar positif dari semua prediksi positif. Recall merupakan rasio prediksi yang benar positif dari semua kasus positif yang sebenarnya.

Apabila hasil dari akurasi, presisi, dan recall mendekati 1, maka akan semakin baik kinerja model yang digunakan

```python
# Lakukan prediksi pada data testing dan buat confusion matrix untuk masing-masing model:

# K-Nearest Neighbors
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)

# Gaussian Naive Bayes
gnb.fit(x_train, y_train)
y_pred_gnb = gnb.predict(x_test)

# Decision Tree Classifier
dtc.fit(x_train, y_train)
y_pred_dtc = dtc.predict(x_test)
```

```python
# Hitung ROC curve dan AUC score untuk masing-masing model:

# K-Nearest Neighbors
cm_knn = confusion_matrix(y_test, y_pred_knn)
fpr_knn, tpr_knn, threshold_knn = roc_curve(y_test, y_pred_knn)
roc_auc_knn = auc(fpr_knn, tpr_knn)

# Gaussian Naive Bayes
cm_gnb = confusion_matrix(y_test, y_pred_gnb)
fpr_gnb, tpr_gnb, threshold_gnb = roc_curve(y_test, y_pred_gnb)
roc_auc_gnb = auc(fpr_gnb, tpr_gnb)

# Decision Tree Classifier
cm_dtc = confusion_matrix(y_test, y_pred_dtc)
fpr_dtc, tpr_dtc, threshold_dtc = roc_curve(y_test, y_pred_dtc)
roc_auc_dtc = auc(fpr_dtc, tpr_dtc)
```

```python
# ROC curve
# Buat plot ROC curve untuk masing-masing model:

plt.plot(fpr_knn, tpr_knn, 'b', label='AUC KNN = %0.2f' % roc_auc_knn)
plt.plot(fpr_gnb, tpr_gnb, 'g', label='AUC GNB = %0.2f' % roc_auc_gnb)
plt.plot(fpr_dtc, tpr_dtc, 'r', label='AUC DTC = %0.2f' % roc_auc_dtc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: cyan' if v else '' for v in is_max]

# Perhitung ROC-AUC
roc_auc_knn = roc_auc_score(y_test, knn.predict(x_test))
roc_auc_gnb = roc_auc_score(y_test, gnb.predict(x_test))
roc_auc_dtc = roc_auc_score(y_test, dtc.predict(x_test))

eval_knn = pd.DataFrame({
    'Model': ['K-Nearest Neighbors'],
    'Accuracy': [np.sum(cm_knn.diagonal())/np.sum(cm_knn)],
    'Precision': [cm_knn[1,1]/np.sum(cm_knn[:,1])],
    'Recall': [cm_knn[1,1]/np.sum(cm_knn[1,:])],
    'F1-Score': [2*cm_knn[1,1]/(2*cm_knn[1,1]+cm_knn[0,1]+cm_knn[1,0])],
    'ROC-AUC' : [roc_auc_knn]
})

eval_gnb = pd.DataFrame({
    'Model': ['Gaussian Naive Bayes'],
    'Accuracy': [np.sum(cm_gnb.diagonal())/np.sum(cm_gnb)],
    'Precision': [cm_gnb[1,1]/np.sum(cm_gnb[:,1])],
    'Recall': [cm_gnb[1,1]/np.sum(cm_gnb[1,:])],
    'F1-Score': [2*cm_gnb[1,1]/(2*cm_gnb[1,1]+cm_gnb[0,1]+cm_gnb[1,0])],
    'ROC-AUC' : [roc_auc_gnb]
})

eval_dtc = pd.DataFrame({
    'Model': ['Decision Tree Classifier'],
    'Accuracy': [np.sum(cm_dtc.diagonal())/np.sum(cm_dtc)],
    'Precision': [cm_dtc[1,1]/np.sum(cm_dtc[:,1])],
    'Recall': [cm_dtc[1,1]/np.sum(cm_dtc[1,:])],
    'F1-Score': [2*cm_dtc[1,1]/(2*cm_dtc[1,1]+cm_dtc[0,1]+cm_dtc[1,0])],
    'ROC-AUC' : [roc_auc_dtc]
})

eval_df = pd.concat([eval_knn, eval_gnb, eval_dtc], ignore_index=True)

eval_df = eval_df.style.apply(highlight_max, subset=['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC'])

eval_df
```
